## Q14. What are the top 10 most popular apps in the dataset? Create a frequency table to identify the apps with the highest number of installs.

```python
df['Installs'].unique()
```

```python
df_copy=df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
```

```python
numeric_features = [feature for feature in df_copy.columns if df_copy[feature].dtype != 'O']
categorical_features = [feature for feature in df_copy.columns if df_copy[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
## Proportion of count data on categorical columns
for col in categorical_features:
    print(df_copy[col].value_counts(normalize=True)*100)   ## normalize gives percentage 
    print('---------------------------')
```

```python
## Top 10 App Categories
category = pd.DataFrame(df_copy['App'].value_counts())        #Dataframe of apps on the basis of category
category.rename(columns = {'App':'Count'},inplace=True)
```

```python
category
```

```python
df_cat_installs = df_copy.groupby(['App'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'App' , data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular App's in Play Store", size = 20)
```

## Q17. A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to launch a new app. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

```python
df_copy.head(2)
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.head(2)
```

```python
df_cat_installs = df_copy.groupby(['Month'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(y = 'Installs' , x= 'Month' , data = df2 )
ax.set_xlabel('No. of Months')
ax.set_ylabel('')
ax.set_title("No. of Installation per Month in Play Store", size = 20)
```

## The most installations happend in the months of 7 and 8 . Therefore the best time to launch a new app in the months of July and August, i.e. 7 & 8.
