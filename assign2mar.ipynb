## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

## A scatter plot is a diagram where each value in the data set is represented by a dot. The Matplotlib module has a method for drawing scatter plots, it needs two arrays of the same length, one for the values of the x-axis, and one for the values of the y-axis.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
## Note: Also add title, xlabel, and ylabel to the plot.
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.xlabel("This is scatter x axis")
plt.ylabel("This is scatter y axis")
plt.title("This is scatter x vs y")
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


## A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories.

```python

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
company
```

```python
profit
```

```python
plt.figure(figsize = (4, 5))
plt.bar(company, profit, color ='maroon')
 
plt.xlabel("company_name")
plt.ylabel("profit")
plt.title("company_profit")
plt.show()
```

```python
plt.figure(figsize = (4, 5))
plt.barh(company, profit, color ='maroon')
 
plt.xlabel("company_name")
plt.ylabel("profit")
plt.title("company_profit")
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


## Box plots are used to show distributions of numeric data values, especially when you want to compare them between multiple groups.
## A Boxplot also called a whisker plot, is a data visualization technique that visualizes the distribution of data from the dataset provided.
## Boxplots are a measure of how well distributed the data in a data set is. It divides the data set into three quartiles. This graph represents the minimum, maximum, median, first quartile and third quartile in the data set.

```python

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)
 
data = [box1, box2]
 
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0,0,1, 1])
 
# Creating plot
bp = ax.boxplot(data)
 
# show plot
plt.show()
```
